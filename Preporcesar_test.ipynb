```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
test = pd.read_csv('test.csv')
```

```python
test_df = pd.DataFrame()
test_df['LotFrontage'] = test.LotFrontage.interpolate()
```

```python
map_d = {'N': 0, 'Y': 1}
test_df['CentralAir'] = test.CentralAir.map(map_d)
```

```python
columnas = ['LotArea', 'YearBuilt', 'GrLivArea', 'BedroomAbvGr', 'KitchenAbvGr', 'Fireplaces', 'GarageArea', 'OpenPorchSF']
for i in columnas:
    test_df[str(i)] = test[str(i)]
```

```python
test.info()
```

```python
test_df
```

```python
test_df['Condition']= np.where(test['Condition1'] != test['Condition2'], 0, 1)
```

```python
test_df['Mean_rank'] = (test.OverallQual + test.OverallCond) / 2
```

```python
test_df['YearRemod'] = np.where(test['YearRemodAdd'] == test['YearBuilt'], 0, 1)
test_df['Exterior'] = np.where(test['Exterior1st'] == test['Exterior2nd'], 0, 1)
test_df['TotalSF'] = test['TotalBsmtSF'] + test['1stFlrSF'] + test['2ndFlrSF']
```

```python
col = ['BsmtFullBath', 'BsmtHalfBath']

for i in col:
    test[str(i)]= test[str(i)].fillna(1)
test_df['TotalBath'] = test['BsmtFullBath'] + test['BsmtHalfBath']/2 + test['FullBath'] + test['HalfBath']/2
```

```python
test_df['Age'] = test.YrSold - test.YearBuilt
test_df['WoodYN'] = np.where(test.WoodDeckSF > 0, 0, 1)
test_df.set_index(test.Id, inplace = True)
```

```python
df = pd.read_csv('Train_data.csv').set_index('Id')
df.drop(['Unnamed: 0'], axis = 1, inplace=True)
df.shape
```
